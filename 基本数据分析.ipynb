#### 首先读入数据文件city_GDP.xls

```python
df = pd.read_excel('data/city_GDP.xls')
df[:5]
```

## 数据格式转换
   - 在做数据分析的时候，原始数据往往会因为各种个各样的原因产生各种数据格式的问题

### 查看格式

```python
df['2018年'].dtype
```

#### 转换格式

```python
#将'2018年'列由'float64'转化为'int'
df['2018年'] = df['2018年'].astype('int')
df[:5]
```
