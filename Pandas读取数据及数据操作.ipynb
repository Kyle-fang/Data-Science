### 缺失值和异常值处理

#### 缺失直处理方法
   - dropna：根据标签中的缺失直进行过滤，删除缺失直
   - fillna：对缺失直进行填充
   - isnull：返回一个布尔对象，判断那些是缺失直
   - notnull：isnull的否定式

#### 判断缺失直

```python
df1[df1['2018年'].isnull()]
```

#### 填充缺失直

```python
df1['2018年'].fillna(1,inplace=True)
df1
```

```python
# 填充数据中所有的缺失直
df2 = df1.fillna('未知')
df2
```

#### 删除缺失直

df.dropna():参数：<br/>
&emsp;how='all':删除全为空值的行或列<br/>
&emsp;inplace='True':覆盖之前的数据<br/>
&emsp;axis=0:选择行或列



```python
df3 = df1.dropna(axis=0)
df3
```

## 处理异常值
   - 异常直，即在数据集中处理存在不合理的值，又称离群点。如：人的年龄为－1，电脑重量为1吨；

```python
df3[df3['2018年']<100]
```

### 对于异常直，一般数量都会很少，在不影响整体数据分布的情况下，直接删除就行了
